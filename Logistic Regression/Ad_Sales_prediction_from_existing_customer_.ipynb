Dividing Input and Output

```python
x=dataset.iloc[:,:-1].values
x
```

```python
y=dataset.iloc[:,-1].values
y
```

Spliting the dataset. training=>0.75 and testing=>0.25, not random selection

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

Feature Scaling-> Giving equal importance to all input columns

x_train has fit_transform because fit() learns the parameter and transform() uses the learned parameter to transform the data

x_test has only transform() because the model model already learned the parameters and only has to transform it

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

Loading the algorithm

```python
from sklearn.linear_model import LogisticRegression
model_load=LogisticRegression()
model_load.fit(x_train,y_train)
```

Predicting the output

```python
y_pred=model_load.predict(x_test)
```

Evaluating the model

```python
from sklearn.metrics import accuracy_score
print(f"The accuracy score is {accuracy_score(y_test,y_pred)*100}")
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
age=int(input("Enter the age:"))
salary=float(input("Enter the salary:"))
df=[[age,salary]]
output=model_load.predict(ss.transform(df))

if output==1:
  print("Customer will buy the product")
else:
  print("Customer will not buy the product")
```
