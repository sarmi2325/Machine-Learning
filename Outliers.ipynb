```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weight-height.csv")
df
```

```python
plt.hist(df.Height,bins=20,rwidth=0.7)
plt.xlabel('Height')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df.Weight,bins=20,rwidth=0.7)
plt.xlabel('Weight')
plt.ylabel('Count')
plt.show()
```

```python
df.describe()
```

```python
Q1 = df.Height.quantile(0.25)
Q3 = df.Height.quantile(0.75)
IQR = Q3-Q1
Q1,Q3,IQR

```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df[(df.Height<lower_limit)|(df.Height>upper_limit)]
```

```python
Height_no_Outliers = df[(df.Height>lower_limit)&(df.Height<upper_limit)]
Height_no_Outliers
```

```python
Q1 = df.Weight.quantile(0.25)
Q3 = df.Weight.quantile(0.75)
IQR = Q3-Q1
Q1,Q3,IQR
```

```python

lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
lower_limit,upper_limit
```

```python
df[(df.Weight<lower_limit)|(df.Weight>upper_limit)]
```

```python
Weight_no_outliers = df[(df.Weight>lower_limit)&(df.Weight<upper_limit)]
Weight_no_outliers
```
